```python
from passlib.hash import pbkdf2_sha256 as pwd_hash
import time
import secrets
import hashlib
import binascii

def hash_password(password: str) -> str:
    # pbkdf2_sha256 handles arbitrary length passwords and salting internally
    return pwd_hash.using(rounds=290000).hash(password)

def verify_password(password: str, hashed: str) -> bool:
    return pwd_hash.verify(password, hashed)

# Demo: hash a weak and a strong password and time verification
weak = "password123"
strong = "Tr0ub4dor&3!$XyZ"  # example of stronger password

h_weak = hash_password(weak)
h_strong = hash_password(strong)

print("Weak hash:", h_weak)
print("Strong hash:", h_strong)

# Timing verification (single verify)
start = time.time()
verify_password(weak, h_weak)
t_weak = time.time() - start

start = time.time()
verify_password(strong, h_strong)
t_strong = time.time() - start

print(f"Verification time (weak): {t_weak:.4f}s")
print(f"Verification time (strong): {t_strong:.4f}s")
```
